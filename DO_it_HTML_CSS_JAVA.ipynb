##    📝 웹은 어떻게 움직일까


*   웹: '하이퍼링크로 연결된 정보망(= 전 세계의 정보들이 지름길로 엮여있다. 그래서 어디든 이동하며 볼 수 있는 거대한 도서관이지. 그 정보들을 창문(브라우저)를 통해 볼 수 있는 것이다.

    웹 사이트에 어울리는 여러 장치와 기능을 만드는 것을 웹 개발이라 한다.





*   네트워크 : 여러 컴퓨터를 연결해 놓은 체계
*   라우터(router) : 여러 컴퓨터를 연결할 떄 중간에 관리자 역할을 해주는 기기
*   인터넷- 라우터를 이용한 네트워크를 끊임없이 연결하여 만든 거대한 네트워크
*   서버 : 네트워크에 연결된 컴퓨터가 인터넷에 공개할 정보를 저장하는 역할
*   클라이언트 컴퓨터 : 서버에 연결해서 정보를 보여달라고 요청하는 컴퓨터(좁은 의미로 브라우저)

*   프론트 앤드 : 사용자 앞에 보이는 영역으로, 웹브라우저 화면에 보이는 것
    - HTML, CSS, 자바스크립트 사용
*   백앤드 : 사용자 뒤에서 보이지 않는 영역
    -  자바, PHP, 파이썬, 자바스크립트 등 사용






*   IP 주소(Internet Protocol address) : 인터넷에 연결되어 있는 여러 사이트를 구분하기 위해 붙이는 주소
*   DNS 서버(Domain Name Server) : 도메인을 IP 주소로 바꿔 주는 역할을 하는 것

* HTTP(Hypertext Transfer Protocol) : 클라이언트와 서버는 사용하는 프로그래밍 언어도 동작하는 방식도 다르지만 ,둘 다 공통으로 인식하는 규칙이 있어 소통할 수 있다. 이 규칙을 HTTP프로토콜이라 함
* HTTP + security -> HTTPS 프로토콜

# 📝웹 개발 시작하기

웹 개발의 기본 - HTML, CSS 자바스크립트

* 웹 문서의 뼈대를 만드는 HTML( 건물의 기초와 벽, 지붕 등의 구조를 만드는 것)
* 웹 문서를 꾸미는 CSS(건물에 색을 칠하고 창문을 디자인하거나 내부 인테리어 작업하는 것)
* 사용자 동작에 반응하는 자바스크립트(건물을 짓고 난 후 엘리베이터를 설치, 에어컨 등 기능 추가하는 것)




# 📝 HTML 기본 문서 만들기



*   HTML(HyperText Markup Language): 웹 문서를 만드는 언어
    * Hypertext: 문서를 서로 연결해주는 링크/ Markup: 웹브라우저에 내용을 보여주는 텍스트, 이미지, 영상 등이 위치를 표시하는 것
    * Tag : 웹 브라우저가 제목, 텍스트, 표를 구별할 수 있도록 각각에 꼬리를 붙인 것,



```
# - <!DOCTYPE html> : HTML 언어로 작성한 웹 문서
- <html> ~ </html>" : 웹 문서의 시작과 끝을 알림
- <head> ~ </head> : 웹 문서의 정보를 저장
- <body> ~ </body> : 웹 브라우저에 보여줄 내용 저장

- <html lang= "ko">: 한국어로 된 문서
- <meta charset="UTF-8> : 화면에 한글로 된내용을 표시할 떄 사용

- <title> ~ </title> : 제목입력
- <body> ~ </body> : 내용 표시

- <p> : paragraph

```



📌 태그
- 태그는 소문자로 씁니다.
- 여는 태그와 닫는 태그, 정확히 입력합니다
- 적당하게 들여쓰기를 합니다.
- 일부 태그는 속성과 함꼐 사용합니다

        - ex. <img src(파일의 경로)="images/salade/jpg" alt(이미지 보조설명) = "레드향", width(너비)="150">
        

🔹 시맨틱 태그(semantic paragraph)
    - 태그 등 이름만으로도 어떤 역할을 하는지 쉽게 알 수 있다.
    - 웹 부라우저가 HTML 코드만 보고도 어느 부분이 제목이고 메뉴인고 본문 내용인지 쉽게 알 수 있기 떄문
    - 문서 구조가 정확히 나눠지므로 PC나 모바일의 웹브라우저와 여르 스마트 기기의 다양한 화면에서 웹 문서를 표현하기가 쉽기 떄문

    <head> : 헤더영역
    <nav> : 내비게이션 영역, 다른 위치로 연결하는 링크
    <main> : 메인 콘텐츠가 들어 있는 태그, 핵심 콘텐츠
    <article> : 웹에서 실제 보여주고 싶은 내용, 독립적인 콘텐츠
    <section> : 몇 개의 콘텐츠를 묶을 떄 사용,콘텐츠 영역
    <aside> : 사이드 바 영역
    <footer> : 웹 문서의 맨 아래쪽, 사이트 제작정보, 저작권 정보, 연락처 등
    <div>: 헤더 내비게이션 푸터 등을 구별할 때 사용

# 📝 웹 문서에 다양한 내용 입력하기

💻 텍스트 입력하기
```
1. <hn> ~ </hn> : 제목을 나타내는 태그
2. <p> ~ </p> : 텍스트 단락을 만드는 태그
3. <br> : 줄을 바꾸는 태그
4. <hr> : 분위기를 전환하는 태그
5. <blockquote> 인용문 </blockquote> : 인용할 떄 쓰는 태그
6. <strong> 굵게 강조할 텍스트 </strong> :화면 낭독기에서 강조하여 읽어줌
7. <b> 굵게 표시할 텍스트 </b> : 단순히 글자만 굵게 표시
8. <em> 이탤릭체로 강조할 텍스트 </em> : 문장에서 특정 부분을 강조하고 싶을 때
9. <i> 이탤릭체로 표시할 텍스트 </i> : 마음속의 생각이나 용어, 관용구 등을 사용하는데 사용
10. <cite> 저작물 제목 </cite> : 책이나 영화, 블로그 같은 저작물의 제목
```



💻 목록 만들기

```
1. 순서 있는 목록
<ol>
    <li> 항목 1 </li>
    <li> 항목 2 </li>
</ol>


2. 순서 없는 목록
<ul>
    <li> 항목 1 </li>
    <li> 항목 2 </li>
</ul>

3. 설명 목록
<dl>
    <dt>이름</dt>
    <dd>값</dd>
</dt>

- dt:description list (전체 컨테이너)
- dt:description term( 제목)
- dd:description data?(내용설명)



```



💻 표 만들기


```
1. 표 만들기
<table>
    <caption> 표 제목 </caption>
</table>

2. 행 만들기 - <tr>태그는 행, <td>태그는 열
<table>
    <tr>
        <td> 1행 1열</td>
        <td> 1행 2열</td>
    </tr>
    <tr>
        <td> 2행 1열</td>
        <td> 2행 2열</td>
    </tr>
</table>


3. 행이나 열을 합치는 rowspan, colspan
<td rowspan = "합칠 셀의 개수">내용</td>
<td colspan = "합칠 셀의 개수">내용</td>


```



💻 이미지 삽입하기


```
1. 이미지 삽입
<img src = "이미지 파일 경로" alt = "대체 텍스트">

<img src="image/salad.jpg" alt = "레드향" width="50%">

2. 웹 요소에 설명 글 붙이기
<figure>
    <img src="image/spring.jpg" alt = "제주의 봄" width="50%">
    <figcaption> 가장 먼저 봄을 맞이하는 제주도</figcaption>
</figure>

```




💻 오디오와 비디오 삽입하기


```
1. 멀티미디어 파일 삽입
<object width="너비" height="높이" data="파일"></object>
# <object> : data 연결
<embed src ="파일 경로" width="너비" height="높이">
# <embed> : source로 연결, 끝태그 없다</embed> 없다!

2. 오디오와 비디오 삽입하기
<audio src="오디오 파일 경로"></audio>
<video src="비디오 파일 경로"></video>

<audio src="medias/spring.mp3" controls></audio>
<video src="medias/salad.mp4" controls width="700"></video>

<audio src="medias/spring.mp3" controls loop></audio>    # loop -> 오디오 파일 계속 반복

<video src="medias/salad.mp4" controls width="700" poster="images/salad.jpg"></video>          # 섬네일 이미지 삽입하기
<video src="medias/salad.mp4" controls width="700" poster="images/salad.jpg" muted autoplay loop></video>    # 오디오를 제거한 후 비디오 자동 재생하기

```



💻 하이퍼링크 삽입하기

```
1. 링크를 삽입하는 <a> 태그와 href 속성
텍스트 링크 만들기
<a href="order.html">주문서 작성하기</a>

이미지 링크 만들기
<a href="order.html"><img src="imgaes/tangerines.jpg" alt="레드향></a>

링크에서 새 탭을 열어 주는 target 속성
<a href="order.html" target="_blank">주문서 작성하기</a>

한 페이지 안에서 점프하는 앵커 만들기
<태그 id="앵커명">내용</태그>
<a href="#앵커명">링크</a>
#앵커명은 유일해야 한다!!!

<a href="#info">레드향</a></li>
<a href="#recipe">레드향 샐러드 레시피</a></li>
.
.
.
<h1 id="#info">레드향</h1>
<h1 id="#recipe">레드향 샐러드 레시피</h1>

```
